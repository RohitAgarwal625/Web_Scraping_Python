```python
import requests
from bs4 import BeautifulSoup
import urllib.parse
import os

# URL for the product page
url = 'https://www.sick.com/in/en/catalog/products/detection-sensors/photoelectric-sensors/w4/wtb4fp-22161120a00/p/p661408?tab=detail'

# Function to download file
def download_file(file_url, file_name):
    response = requests.get(file_url, stream=True)
    with open(file_name, 'wb') as file:
        for chunk in response.iter_content(chunk_size=1024):
            if chunk:
                file.write(chunk)

# Fetch the HTML content of the page
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
print(response.content)

# Extract the table data
product_details = {}
table = soup.find('table', {'class': 'table'})
if table:
    rows = table.find_all('tr')
    for row in rows:
        cols = row.find_all('td')
        if len(cols) == 2:
            key = cols[0].text.strip()
            value = cols[1].text.strip()
            product_details[key] = value

# Display the extracted product details
print("Product Details:")
for key, value in product_details.items():
    print(f"{key}: {value}")

# Extract the product image
image_tag = soup.find('img', {'class': 'product-image'})
if image_tag:
    image_url = urllib.parse.urljoin(url, image_tag['src'])
    image_name = os.path.basename(image_tag['src'])
    print(f"\nDownloading Product Image: {image_url}")
    download_file(image_url, image_name)
    print(f"Product image downloaded as {image_name}")

# Extract the datasheet URL
datasheet_tag = soup.find('a', {'title': 'Download product datasheet'})
if datasheet_tag:
    datasheet_url = urllib.parse.urljoin(url, datasheet_tag['href'])
    datasheet_name = "product_datasheet.pdf"
    print(f"\nDownloading Product Datasheet: {datasheet_url}")
    download_file(datasheet_url, datasheet_name)
    print(f"Product datasheet downloaded as {datasheet_name}")

print("\nTask completed.")
```
